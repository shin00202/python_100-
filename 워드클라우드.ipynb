### 텍스트 가져오기

```python
text = open(r'C:\Users\user\Desktop\p1\thelittleprince.txt').read()
```

```python
text
```

```python
img = r'C:\Users\user\Desktop\p1\little.png'
```

```python
mg=Image.open(img)
```

```python
mg
```

```python
lp_mask=np.array(mg)
lp_mask
```

```python
lp_wc=WordCloud(background_color = 'white', 
                mask=lp_mask, 
                max_font_size=50, 
                contour_width=5, 
                contour_color='green').generate(text)
```

```python
lp_wc.words_
```

```python
lp_wc.to_image()
```

```python
lp_wc.to_file()
```
