```python
t = np.arange(0,12,0,5)
y1=np.sin(t)
y2=np.cos(t)
plt.grid()
plt.plot(t, y1, 'y--', marker = 'o')
plt.plot(t, y2, 'r--', marker = '*')
plt.xlabel('Time')
plt.ylabel('Amp')
plt.title('SinCos')
plt.legend(['Sin'.'Cos'])
```
